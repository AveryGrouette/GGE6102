```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\Avery\OneDrive\Documents\GitHub\GGE6102\Data_Sept_F180_Combined.csv')

#starting positions

X = df.iloc[2,3]
Y = df.iloc[2,4]
H = df.iloc[2,5]
output = pd.DataFrame(columns=['X', 'Y', 'H'])

for itx, row in df.iterrows():
    X = X + float(row['Ve(m/s)'])
    Y = Y + float(row['Vn (m/s)'])
    H = H + float(row['Vu(m/s)'])
    output = output.append({"X": X, "Y": Y, "H": H}, ignore_index = True)

output.to_csv('INS.csv')

```
